```python
#Branching with if statements 
i=6
if i<5:
    print("You are allowed to enter for free")
    
if i>=5:
    print ('You must pay to enter')
    
    
```

```python
#Making this code more efficent with the else statement 
import random 
i=random.randint(1,100)
print(i)
if i<50:
    print("You are allowed to enter for free")
    
else: 
    print('You must pay to enter')
```

```python
#elif statement 
i=random.randint(1,100)
print(i)
if i<33:
    print("Go have fun ")
elif (33<=i<=66):
    print('Go to work')
    
else: 
    print('Relax')
```
