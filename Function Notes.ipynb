```python
#Functions 
#Function to calcuate the xth Catalan number
import operator as op
from functools import reduce

def ncr(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer // denom

def Cat(x):
    y=1/(x+1)*ncr(2*x,x)
    
    return y

ncr(4,2)
Cat(3)
    
```
